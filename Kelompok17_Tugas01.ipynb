### No. 4

Reinforcement Learning yaitu model belajar untuk melihat situasi dan menentukan aksi untuk memperoleh hasil terbaik.
Karakteristik Reinforcement Learning sendiri adalah ketika kita bermain game. 
Masalah Riil => Pada metode Machine Learning jenis ini RL, agen akan belajar mendapatkan reward sebesar-besarnya (long term reward). Reward diperoleh berdasarkan aksi yang dilakukan pada suatu kondisi (state). Setia aksi yang dilakukan nantinya akan menciptakan perubahan state yang akan mempengaruhi reward selanjutnya. Tujuan akhir dari reinforcement learning dalam permainan game adalah menang.

Jadi, kapan kita akan menggunakan Reinforcement Learning? Reinforcement Learning akan digunakan ketika terjadi interaksi antara kita sendiri sebagai pembuat dan mesin, long-term reward, hingga Decision Making (Pengambil Keputusan). Pada Reinforcement Learning, sulit untuk mendapatkan data yang “benar” untuk setiap kondisi. Namun, Reinforcement Learning memiliki referensi “benar/salah” berdasarkan reward yang telah diperoleh.
